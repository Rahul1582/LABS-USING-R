# Now cleaning the sentences..

```python
# Lowercase all characters
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: x.lower())
new_df['english_sentence']=new_df['english_sentence'].apply(lambda y: y.lower())
```

```python
import  re
import string
from string import digits
```

```python
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: re.sub("'", '', x))
new_df['english_sentence']=new_df['english_sentence'].apply(lambda x: re.sub("'", '', x))
exclude = set(string.punctuation) # Set of all special characters
# Remove all the special characters
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
new_df['english_sentence']=new_df['english_sentence'].apply(lambda x: ''.join(ch for ch in x if ch not in exclude))
```

```python
# Remove all numbers from text
remove_digits = str.maketrans('', '', digits)
new_df['english_sentence']=new_df['english_sentence'].apply(lambda x: x.translate(remove_digits))
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: x.translate(remove_digits))

new_df['hindi_sentence'] = new_df['hindi_sentence'].apply(lambda x: re.sub("[२३०८१५७९४६]", "", x))

# Remove extra spaces
new_df['english_sentence']=new_df['english_sentence'].apply(lambda x: x.strip())
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: x.strip())
new_df['english_sentence']=new_df['english_sentence'].apply(lambda x: re.sub(" +", " ", x))
new_df['hindi_sentence']=new_df['hindi_sentence'].apply(lambda x: re.sub(" +", " ", x))
```

# Now calculating the english and hindi sentence length..

```python
senlen1=[]
```

```python
for i in new_df['english_sentence']:
    res = len(i.split()) 
    senlen1.append(res)
```

```python
new_df['engsentlen']=senlen1
```

```python
new_df=new_df[new_df['engsentlen']<30]
new_df=new_df[new_df['hinsenlen']<30]
```

# Texts to Sequence conversion-->

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(new_df['english_sentence'].values)
english=tokenizer.texts_to_sequences(new_df['english_sentence'].values)
eng_vocab_size = len(tokenizer.word_index) + 1
```

```python
print('English Vocabulary Size: %d' % eng_vocab_size)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(new_df['hindi_sentence'].values)
hindi=tokenizer.texts_to_sequences(new_df['hindi_sentence'].values)
hin_vocab_size = len(tokenizer.word_index) + 1
```

```python
print('Hindi Vocabulary Size: %d' % hin_vocab_size)
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
english = pad_sequences(english, maxlen=max_length_eng, padding='post')
```

```python
hindi = pad_sequences(hindi, maxlen=max_length_hin, padding='post')
```

# Now splitting the data into training and testing..

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(english, hindi, test_size=0.2, random_state = 42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, RepeatVector,Bidirectional
```

```python
model = Sequential()
model.add(Embedding(input_dim=25867,output_dim=90,input_length=max_length_eng))
model.add(Dropout(0.2))
model.add(LSTM(units=512,return_sequences=False))
model.add(RepeatVector(29))
model.add(LSTM(units=512,return_sequences=True))
model.add(Dense(29710,activation='softmax'))
```

```python
from keras.optimizers import Adam
```

```python
learning_rate = 0.005
    
model.compile(loss ='sparse_categorical_crossentropy', 
                 optimizer = 'adam', 
                 metrics = ['accuracy'])
```
