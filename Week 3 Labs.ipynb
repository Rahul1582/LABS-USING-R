```r
lm.fit=lm(medv~lstat,data=Boston)
attach(Boston)
lm.fit=lm(medv~lstat)
```

```r
lm.fit
```

```r
summary(lm.fit)
names(lm.fit)
coef(lm.fit)
confint(lm.fit)
```

```r
predict(lm.fit,data.frame)
```

```r
plot(lstat,medv)
abline(lm.fit)
```

```r
predict(lm.fit,data.frame(lstat=c(2,4,6)),interval="confidence")
predict(lm.fit,data.frame(lstat=c(3,5,7)),interval='prediction')
plot(lstat,medv)
abline(lm.fit,lwd=3)
abline(lm.fit,lwd=3,col='red')
plot(lstat,medv,pch=20)
plot(1:20,1:20,pch=1:20,xlab='lstat',ylab='medv')
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
plot(predict(lm.fit),residuals(lm.fit))
plot(predict(lm.fit),rstudent(lm.fit))
plot(hatvalues(lm.fit))
which.max(hatvalues(lm.fit))
```

```r
lm.fit=lm(medv~lstat+age,data=Boston)
summary(lm.fit)
```

```r
lm.fit=lm(medv~.,data=Boston)
summary(lm.fit)
summary(lm.fit)$sigma
```

```r
lm.fitt=lm(medv~.-age-rad-rm,data=Boston)
summary(lm.fitt)
```

```r
summary(lm(medv~lstat*age,data=Boston))
```

```r
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
lm.fit=lm(medv~lstat)
anova(lm.fit,lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```
